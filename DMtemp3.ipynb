```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bc = "companydata.csv"
df = pd.read_csv(bc, header = 0)
df
```

```python
test = df.replace('?', np.NaN)
test
```

```python
#gotta change dem objects to floats man
test.dtypes.eq(object)
c = test.columns[test.dtypes.eq(object)]
test[c] = test[c].apply(pd.to_numeric, errors='coerce', axis=0)
test.dtypes 
```

```python
#filling missing values instead of deleting them
test2=test.fillna(test.mean())
```

```python
test2
```

```python
from sklearn import preprocessing

x = test2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
test2 = pd.DataFrame(x_scaled)
```

```python
test2
```

```python
#tree classifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
#x65 now reffered to 64 after normalizing
X=test2.drop(64,axis=1)
```

```python
y=test2[64]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='entropy')
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```
